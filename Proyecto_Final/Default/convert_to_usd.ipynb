```python
# Define the exchange rates for TWD to USD
exchange_rates = {
    'enero': 0.03062,
    'febrero': 0.03083,
    'marzo': 0.03125,
    'abril': 0.03106,
    'mayo': 0.03114,
    'junio': 0.03118,
    'julio': 0.03134,
    'agosto': 0.03132,
    'septiembre': 0.03141,
    'octubre': 0.03107,
    'noviembre': 0.03116,
    'diciembre': 0.03125,
    'annual': 0.031
}
# Promedio de tipos de cambio de TWD a USD para 2005
average_exchange_rate_2005 = 0.031  # Reemplaza con el promedio que calculaste


def convert_currency(row):
    # Convert relevant columns using the exchange rates
    row['LIMITE_CREDITO'] *= exchange_rates['annual']

    row['MONTO_FACTURA_1'] *= exchange_rates['septiembre']
    row['MONTO_FACTURA_2'] *= exchange_rates['agosto']
    row['MONTO_FACTURA_3'] *= exchange_rates['julio']
    row['MONTO_FACTURA_4'] *= exchange_rates['junio']
    row['MONTO_FACTURA_5'] *= exchange_rates['mayo']
    row['MONTO_FACTURA_6'] *= exchange_rates['abril']
    
    row['MONTO_PAGO_1'] *= exchange_rates['septiembre']
    row['MONTO_PAGO_2'] *= exchange_rates['agosto']
    row['MONTO_PAGO_3'] *= exchange_rates['julio']
    row['MONTO_PAGO_4'] *= exchange_rates['junio']
    row['MONTO_PAGO_5'] *= exchange_rates['mayo']
    row['MONTO_PAGO_6'] *= exchange_rates['abril']
    
    return row

# Apply the conversion function to the DataFrame
df_converted = df.apply(convert_currency, axis=1)
```

### Separate test set

```python
from sklearn.model_selection import train_test_split

df = pd.read_excel('/home/linar/Desktop/ML/Clases/i302/Proyecto_Final/Default/finanzas_usd.xlsx')

# Define bins and labels
bins = [0, 310, 1550, 4340, 7440, 31000]
labels = ['Muy Bajo', 'Bajo', 'Medio-Bajo', 'Medio-Alto', 'Alto']

# Create a new column for the credit limit category
df['credit_limit_category'] = pd.cut(df['LIMITE_CREDITO'], bins=bins, labels=labels, right=False)

# Separate rows with missing values
missing_values_df = df[df['credit_limit_category'].isnull()]

# Drop rows with missing values from the main DataFrame
df = df.dropna(subset=['credit_limit_category'])

# Perform stratified sampling to create a test set with 10% of the data from non-missing values
train_df, test_df = train_test_split(df, test_size=0.1, stratify=df['credit_limit_category'], random_state=42)

# Combine the training set with the missing values DataFrame
train_df = pd.concat([train_df, missing_values_df], ignore_index=True)

# Show the results
print("Train set distribution:")
print(train_df['credit_limit_category'].value_counts(normalize=True))

print("\nTest set distribution:")
print(test_df['credit_limit_category'].value_counts(normalize=True))
```

```python
# Shuffle the training set
train_df = train_df.sample(frac=1, random_state=42).reset_index(drop=True)

# Remove the credit limit category column before saving
train_df = train_df.drop(columns=['credit_limit_category'])
test_df = test_df.drop(columns=['credit_limit_category'])

# Save the resulting DataFrames to new Excel files or CSV files
train_df.to_excel('/home/linar/Desktop/ML/Clases/i302/Proyecto_Final/Default/finanzas_dev.xlsx', index=False)
test_df.to_excel('/home/linar/Desktop/ML/Clases/i302/Proyecto_Final/Default/finanzas_test.xlsx', index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of seaborn
sns.set_theme(style='whitegrid')

# Define the columns to plot
columns_to_plot = ['LIMITE_CREDITO', 'GENERO', 'EDUCACION', 'ESTADO_CIVIL', 'EDAD']

# Create a figure for the plots
plt.figure(figsize=(15, 10))

# Loop through each column and create subplots
for i, column in enumerate(columns_to_plot, 1):
    plt.subplot(3, 2, i)  # Create a grid of 3x2 subplots
    sns.histplot(train_df[column], color='blue', label='Train', stat='density', kde=True, bins=30, alpha=0.5)
    sns.histplot(test_df[column], color='orange', label='Test', stat='density', kde=True, bins=30, alpha=0.5)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Density')
    plt.legend()

# Adjust layout
plt.tight_layout()
plt.show()
```

### Double check: Compare distributions of real dataset and new usd dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
usd_df = pd.read_excel('/home/linar/Desktop/ML/Clases/i302/Proyecto_Final/Default/finanzas_usd.xlsx')
original_df = pd.read_excel('/home/linar/Desktop/ML/Clases/i302/Proyecto_Final/Default/finanzas.xls')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style of seaborn
sns.set_theme(style='whitegrid')

# Define the columns to plot (these are the monetary columns that were converted)
columns_to_compare = ['LIMITE_CREDITO', 'MONTO_FACTURA_1', 'MONTO_FACTURA_2', 'MONTO_FACTURA_3', 
                      'MONTO_FACTURA_4', 'MONTO_FACTURA_5', 'MONTO_FACTURA_6', 
                      'MONTO_PAGO_1', 'MONTO_PAGO_2', 'MONTO_PAGO_3', 
                      'MONTO_PAGO_4', 'MONTO_PAGO_5', 'MONTO_PAGO_6']

# Create a figure for the plots
plt.figure(figsize=(15, 20))

# Loop through each column and create subplots
for i, column in enumerate(columns_to_compare, 1):
    plt.subplot(7, 2, i)  # Create a grid of 6x2 subplots
    sns.kdeplot((original_df[column]*0.031).dropna(), color='blue', label='Original (TWD)', fill=True, alpha=0.3)
    sns.kdeplot(usd_df[column].dropna(), color='red', label='Converted (USD)', fill=True, alpha=0.3)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Density')
    plt.legend()

# Adjust layout
plt.tight_layout()
plt.show()

# Step 2: Check the mean and standard deviation before and after conversion
for column in columns_to_compare:
    original_mean = (original_df[column] * 0.031).dropna().mean()
    original_std = (original_df[column] * 0.031).dropna().std()
    converted_mean = usd_df[column].dropna().mean()
    converted_std = usd_df[column].dropna().std()
    print(f"{column} - Original mean (TWD): {original_mean:.2f}, Converted mean (USD): {converted_mean:.2f}")
    print(f"{column} - Original std (TWD): {original_std:.2f}, Converted std (USD): {converted_std:.2f}")
```
