## Ejemplo 1
> El etilenglicol (CH2OHCH2OH) se suele mezclar con agua para proporcionar soluciones anticongelantes para los sistemas de refrigeraci칩n de los motores de los autos. A medida que se agrega glicol al agua, el punto de congelaci칩n de la soluci칩n cae r치pidamente. Sin embargo, a medida que se acerca al glicol puro, el punto de congelaci칩n aumenta r치pidamente.

| Glicol (% de vol.)        | 0 | 10   | 20   | 30    | 40    | 50    | 60    | 80  | 90  | 100   |
|---------------------------|---|------|------|-------|-------|-------|-------|-----|-----|-------|
| Punto de congelaci칩n (춿C) | 0 | -3.4 | -7.9 | -13.7 | -23.5 | -36.8 | -52.8 | -46 | -30 | -12.8 |

>La Tabla presenta datos sobre el punto de congelaci칩n versus el porcentaje de glicol. Las soluciones anticongelantes t칤picas son glicol al 50 % y brindan protecci칩n hasta -36,8 춿C; aunque, debido a la formaci칩n de aguanieve.

## Graficando m치s de una curva en un mismo gr치fico

Vamos a trabajar con este ejemplo:

> El uso de soluciones de sal (NaCl) para descongelar las superficies de las rutas ha sido reemplazado en las 칰ltimas d칠cadas por cloruro de magnesio (MgCl) para reducir la corrosi칩n del acero en veh칤culos y otras estructuras (aunque el cloruro de magnesio tiene un efecto corrosivo adverso en el aluminio). Las propiedades de estas soluciones son de inter칠s y la siguiente tabla presenta las densidades de estas soluciones a 0춿C:

| Porcentaje | NaCl | MgCl |
| --- | --- | --- |
| 2 | 1.01509 | 1.0168 |
| 4 | 1.03038 |  1.0338 |
| 8 | 1.06121 | 1.0683 |
| 12 | 1.09244 | 1.1035 |
| 16 | 1.12419 | 1.1395 |
| 20 | 1.15663 | 1.1764 |

Creamos un gr치ficos de lineas con marcadores para ambos conjuntos de datos:

```python
Conc = [2, 4, 8, 12, 16, 20]
NaCl = [1.01509, 1.03038, 1.06121, 1.09244, 1.12419, 1.15663]
MgCl = [1.0168, 1.0338, 1.0683, 1.1035, 1.1395, 1.1764]

plt.plot(Conc, NaCl, marker='o', c='r', mec='r', mfc='w')
plt.plot(Conc, MgCl, marker='s', c='b', ls='--', mec='b', mfc='w')
plt.grid()
plt.xlabel('Concentracion - wt%')
plt.ylabel('Densidad - gm/cc')
plt.title('Densidad de soluciones con sal y cloruro de sodio');
```

Podemos ver que usamos un comando de **plot** separado para cada serie.

Un problema inmediato es que no es evidente cu치l de las dos curvas es para NaCl y cu치l es para MgCl. Necesitamos una leyenda. Esto se logra agregando el comando **legend** y las especificaciones de **label** en los comandos de **plot**:

```python
plt.plot(Conc, NaCl, marker='o', c='r', mec='r', mfc='w', label="NaCl")
plt.plot(Conc, MgCl, marker='s', c='b', ls='--', mec='b', mfc='w', label="MgCl")
plt.legend()
plt.grid()
plt.xlabel('Concentracion - wt%')
plt.ylabel('Densidad - gm/cc')
plt.title('Densidad de soluciones con sal y cloruro de sodio');
```

Ahora, est치 claro que curva es cu치l. El comando *legend* coloca autom치ticamente la leyenda de forma que evite lo m치s posible tapar las gr치ficas. Es posible forzar la ubicaci칩n de la leyenda agregando una especificaci칩n de ubicaci칩n al comando de leyenda:

```python
plt.plot(Conc, NaCl, marker='o', c='r', mec='r', mfc='w', label="NaCl")
plt.plot(Conc, MgCl, marker='s', c='b', ls='--', mec='b', mfc='w', label="MgCl")
plt.legend(loc="lower right")
plt.grid()
plt.xlabel('Concentracion - wt%')
plt.ylabel('Densidad - gm/cc')
plt.title('Densidad de soluciones con sal y cloruro de sodio');
```

Como indica el comando, la leyenda ahora se colocar치 en la parte inferior derecha del gr치fico.

Hay momentos en los que no estamos satisfechos con las escalas de los ejes adoptadas autom치ticamente por **pyplot**. Es posible modificar los l칤mites de los ejes y los intervalos (**ticks**) a nuestra elecci칩n. Por ejemplo, en la figura anterior, podr칤amos preferir que el **eje x** tenga l칤mites de 0 y 22 con intervalos de dos unidades, y el **eje y** con l칤mites de 1.0 a 1.2 con intervalos de 0.02:

```python
plt.plot(Conc, NaCl, marker='o', c='r', mec='r', mfc='w', label="NaCl")
plt.plot(Conc, MgCl, marker='s', c='b', ls='--', mec='b', mfc='w', label="MgCl")
plt.legend(loc="lower right")

plt.grid()
plt.xlim(0., 22.)
plt.ylim(1., 1.2)
plt.xticks(np.arange(0,24,2))
plt.yticks(np.arange(1,1.22,0.02))

plt.xlabel('Concentracion - wt%')
plt.ylabel('Densidad - gm/cc')
plt.title('Densidad de soluciones con sal y cloruro de sodio');
```

Las funciones **xlim** e **ylim** establecen los l칤mites del eje x e y, si se desean intervalos espec칤ficos, se pueden usar las funciones **xticks** e **yticks**.

Observe c칩mo la funci칩n **arange** usa un l칤mite un paso m치s all치 de lo deseado. Alternativamente, se puede proporcionar una lista de valores.

Hay ocasiones en las que una trama de m치s de una serie no queda bien porque las series tienen distintas magnitudes. Una soluci칩n a este problema es proporcionar otro eje vertical en el lado derecho de la gr치fica. El siguiente ejemplo ilustra c칩mo lograr esto con Matplotlib:

> Hemos recopilado una semana de datos meteorol칩gicos (finales de octubre a principios de noviembre). Esto representa m치s de 20 puntos de datos. Los datos seleccionados son temperatura (춿C) y humedad relativa (%). Si graficamos estos datos obtenemos:

```python
sample = np.arange(15)
temperature = [10, 20, 3, 10, 18, 5, 11, 19, 3, 10, 18, 5, 11, 19, 1]
humidity = [95, 90, 80, 93, 75, 80, 55, 20, 50, 19, 81, 77, 65, 82, 79]

plt.plot(sample, temperature, c='r', label='Temperatura')
plt.plot(sample, humidity, c='b', ls='--', label='Humedad')
plt.grid()
plt.xlabel('Numero de muestra')
plt.ylabel('Temperatura [춿C] y Humedad relativa [%]')
plt.title('Datos del clima')
plt.legend(loc='center left');
```

Con la leyenda y los estilos de l칤nea, podemos diferenciar las dos curvas. Sin embargo, la humedad var칤a a lo largo del rango del eje vertical; mientras que la temperatura se comprime en la parte inferior. Podemos modificar el c칩digo para colocar un eje en el lado derecho de la gr치fica y usarlo para una de las series usando la funci칩n **twinx**:


```python
plt.plot(sample, temperature, c='r', label='Temperatura')
plt.title('Datos del clima')
plt.grid()

plt.xlabel('Numero de muestra')
plt.ylabel('Temperatura [춿C]')


plt.twinx()
plt.plot(sample, humidity, c='b', ls='--', label='Humedad')
plt.ylabel('Humedad relativa [%]');
```

Las dos curvas cubren ahora una mayor parte del rango vertical. Un elemento que falta aqu칤 es la leyenda. Como se ilustra a continuaci칩n, resulta que no es tan trivial en Python y pyplot agregar la leyenda. El siguiente c칩digo de Python se puede usar para incluir una leyenda en el gr치fico con ejes gemelos.



```python
curve1 = plt.plot(sample, temperature, c='r', label='Temperatura')
plt.title('Datos del clima')
plt.grid()

plt.xlabel('Numero de muestra')
plt.ylabel('Temperatura [춿C]')


plt.twinx()
curve2 = plt.plot(sample, humidity, c='b', ls='--', label='Humedad')
plt.ylabel('Humedad relativa [%]');

# Hacemos esto para poder mostrar los textos de las dos curvas
curves = curve1 + curve2
labels = []
for curve in curves:
  labels.append(curve.get_label())
plt.legend(curves, labels, loc="lower left");
```

El resultado de la funci칩n de trazado se asigna a un objeto, **curve1** o **curve2**. Estos objetos se combinan en la curva de objeto. Luego, las etiquetas de leyenda para cada curva se agregan a un objeto de etiquetas, que se usa en el comando de leyenda.

Los datos meteorol칩gicos graficados presentados pueden no ser satisfactorios para algunos  espectadores debido a la superposici칩n de las dos curvas. Esto podr칤a solucionarse utilizando diferentes colores para las curvas si el color es aceptable como caracter칤stica. Otro enfoque es usar m치s de una gr치fica separadas apiladas verticalmente. Esto es lo que vemos a continuaci칩n.

## Usando objetos de figura

El uso de objetos de figura proporciona flexibilidad adicional en la creaci칩n de gr치ficos utilizando **Matplotlib** y **pyplot**. Podemos crear varios objetos de curvas dentro en un script y podemos incluir varias gr치ficas en una sola figura. Un escenario donde se desean m칰ltiples gr치ficos es donde queremos un gr치fico de dos funciones versus una variable independiente, y luego tambi칠n necesitamos un gr치fico de fase donde los dos valores de funci칩n se trazan uno contra el otro con la variable independiente como par치metro de ejecuci칩n. Esto se logra con el uso repetido del comando figura, como se muestra en el siguiente script:

```python
x = np.linspace(0,np.pi, 100)
f1 = np.sin(x)
f2 = np.cos(x)

plt.figure()
plt.plot(x, f1, c='r', label='sen(x)')
plt.plot(x, f2, c='b', ls='--', label='cos(x)')
plt.grid()
plt.xlabel('x')
plt.ylabel('sen(x) y cos(x)')
plt.title('Funciones trigonometricas')
plt.legend()

plt.figure()
plt.plot(f1, f2, c='m')
plt.grid()
plt.xlabel('sen(x)')
plt.ylabel('cos(x)')
plt.title('cos(x) vs sen(x) para x = 0 a pi');
```

Las dos figuras creadas se muestran una abajo de la otra. Estas son figuras independientes y se pueden organizar en un documento como se prefiera.

Se puede ver que crear varias figuras es simple: solo use el comando **figure** antes de cada una.

Si deseamos crear una estructura de gr치ficos dentro de una sola figura, esto requiere el uso de objetos de ejes y la especificaci칩n de la estructura: cu치ntas gr치ficas hay en cada fila y cu치ntas filas. Apliquemos este m칠todo a las dos gr치ficas de funciones trigonom칠tricas anteriores. A continuaci칩n, creamos un objeto de **figura** llamado **fig**, y luego agregamos objetos de ejes, **ax1 y ax2**, usando el m칠todo **add_subplot**.




```python
fig = plt.figure()

ax1 = fig.add_subplot(121)
ax1.plot(x, f1, c='r', label='sen(x)')
ax1.plot(x, f2, c='b', ls='--', label='cos(x)')
ax1.grid()
ax1.set_xlabel('x')
ax1.set_ylabel('sen(x) y cos(x)')
ax1.set_title('Funciones trigonometricas')
ax1.legend()

ax2 = fig.add_subplot(122)
ax2.plot(f1, f2, c='m')
ax2.grid()
ax2.set_xlabel('sen(x)')
ax2.set_ylabel('cos(x)')
ax2.set_title('cos(x) vs sen(x) para x = 0 a pi');
```

El arreglo especificado, **121** o **122**, es 1 fila y 2 columnas con el 칰ltimo d칤gito que indica en cual subparcela en la figura, contando a traves de las filas se va a poner el grafico.

Luego, en lugar de usar **plt** para **plot** y otros comandos, usamos los objetos de ejes (**ax1** y **ax2**. Adem치s, tenga en cuenta que los comandos anteriores, como **xlabel**, ahora son **set_xlabel**.

Un problema "cosm칠tico" evidente de inmediato en la figura anterior es que las dos gr치ficas se amontonan y se superponen. Hay dos remedios disponibles para este fen칩meno com칰n. El primero es agregar espacio entre las gr치ficas con la especificaci칩n **wspace**:

```python
fig = plt.figure()

ax1 = fig.add_subplot(121)
ax1.plot(x, f1, c='r', label='sen(x)')
ax1.plot(x, f2, c='b', ls='--', label='cos(x)')
ax1.grid()
ax1.set_xlabel('x')
ax1.set_ylabel('sen(x) y cos(x)')
ax1.set_title('Funciones trigonometricas')
ax1.legend()

ax2 = fig.add_subplot(122)
ax2.plot(f1, f2, c='m')
ax2.grid()
ax2.set_xlabel('sen(x)')
ax2.set_ylabel('cos(x)')
ax2.set_title('cos(x) vs sen(x) para x = 0 a pi')

fig.subplots_adjust(wspace=0.5);
```

Si queremos que **matplotlib** se encargue de acomodar sola los margenes, podemos usar el comando **tight_layout()**:

```python
fig = plt.figure()

ax1 = fig.add_subplot(121)
ax1.plot(x, f1, c='r', label='sen(x)')
ax1.plot(x, f2, c='b', ls='--', label='cos(x)')
ax1.grid()
ax1.set_xlabel('x')
ax1.set_ylabel('sen(x) y cos(x)')
ax1.set_title('Funciones trigonometricas')
ax1.legend()

ax2 = fig.add_subplot(122)
ax2.plot(f1, f2, c='m')
ax2.grid()
ax2.set_xlabel('sen(x)')
ax2.set_ylabel('cos(x)')
ax2.set_title('cos(x) vs sen(x) para x = 0 a pi')

fig.tight_layout();
```

Tal vez, no nos gusta que las dos gr치ficas est칠n comprimidas horizontalmente. Esto se puede abordar cambiando la relaci칩n de aspecto (ancho, alto) de la ventana de la figura modificando la declaraci칩n de **figure** como:

```python
fig = plt.figure(figsize=(8,3))

ax1 = fig.add_subplot(121)
ax1.plot(x, f1, c='r', label='sen(x)')
ax1.plot(x, f2, c='b', ls='--', label='cos(x)')
ax1.grid()
ax1.set_xlabel('x')
ax1.set_ylabel('sen(x) y cos(x)')
ax1.set_title('Funciones trigonometricas')
ax1.legend()

ax2 = fig.add_subplot(122)
ax2.plot(f1, f2, c='m')
ax2.grid()
ax2.set_xlabel('sen(x)')
ax2.set_ylabel('cos(x)')
ax2.set_title('cos(x) vs sen(x) para x = 0 a pi')

fig.tight_layout();
```

## Creando otros gr치ficos de inter칠s

Un estilo de gr치fico que algunos prefieren usar en ciertas situaciones es el **stem plot**. Utiliza marcadores y conecta una l칤nea vertical o "tallo" desde la abscisa a cada marcador.

Graficamos el primer gr치fico usando este tipo de gr치fico:

```python
x = np.linspace(0,5)
f = np.cos(x)*np.cosh(x) - 1

plt.stem(x, f, linefmt='b-', markerfmt='rd', basefmt='k--')

plt.grid() # Agrega una grilla
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gr치fica de cos(x)*cosh(x) - 1");
```

En el comando **stem**, las especificaciones son l칤neas verticales de color azul solidas (**b-**) y los marcadores como diamantes rojos (**rd**). La l칤nea de base tambi칠n es de color negro punteado.

Hay ocasiones en las que queremos usar ejes logar칤tmicos, ya sea x, y o ambos. Esto suele surgir cuando el rango de valores abarca varios ordenes de magnitud, y es de inter칠s el comportamiento de la trama en cada orden.

Analicemos esto con el siguiente ejemplo:

> Si consideramos el aire h칰medo a una temperatura y presi칩n dadas, la presi칩n de vapor asociada al agua es la fracci칩n de la presi칩n total correspondiente a la fracci칩n de la mezcla aire-agua que corresponde al agua. Esa fracci칩n es molar y no en masa. En t칠rminos de una f칩rmula, eso ser칤a
>
> $$ P^V_w = Py_w $$
>
> Donde
> - $P^V_w$: Presion del vapor de agua

> Los par치metros de la ecuaci칩n de Antoine para el agua (para la temperatura en 춿C y la presi칩n en kPa), son:

> Queremos desarrollar un gr치fico de presi칩n de vapor versus temperatura que incluya tanto los datos como la curva de la ecuaci칩n de Antoine.

```python
TC = np.array([-17.8, -12.2, -6.7, -1.1, 4.4, 10.0, 15.6, 21.1, 26.7, 32.2,
               37.8, 43.3, 48.9, 54.4, 60.0, 65.6, 71.1, 76.7, 82.2, 87.8, 93.3,
               98.9, 100.0])

PV = np.array([0.1275, 0.2128, 0.3479, 0.5078, 0.8386, 1.227, 1.7659, 2.5023,
               3.4946, 4.8136, 6.5016, 8.7911, 11.6712, 15.3276, 19.921,
               25.6384, 32.692, 41.319, 51.77883, 64.3993, 79.4718, 97.3781,
               101.3289])

A = 7.23325
B = 1750.3
C = 235.

PA = 10**(A-B/(TC+C))

plt.plot(TC, PA, c='r', lw=1.0, label='Ecuaci칩n de Antoine')
plt.scatter(TC, PV, c='w', marker='s', edgecolor="k", label='Tabla de vapor')
plt.grid()
plt.xlabel('Temperatura [춿C]')
plt.ylabel('Presi칩n de vapor [kPa]')
plt.title('Presi칩n de vapor de agua versus temperatura')
plt.legend();
```
