Pandas es una herramienta de manipulación y análisis de datos de código abierto, rápida, potente, flexible y fácil de usar, construida sobre el lenguaje de programación Python.

Pandas provee dos estructuras de datos:
- [**Series**](https://pandas.pydata.org/docs/reference/api/pandas.Series.html#pandas.Series): Es una array unidimensional de cualquier tipo de dato.
- [**DataFrame**](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.html#pandas.DataFrame): Es una estructura dos dimensional tal como una tabla con filas y columnas. Esta es la estructura principal de Pandas, nos permite manipular datos estructurados muy facilmente.

### Importar biblioteca

Usando el alias **pd** para ahorrarnos trabajo cuando llamamemos a la libreria  en el script.

```python
import pandas as pd
```

### Series

```python
# Definición de una Serie
serie1 = pd.Series([17,5,7,5,67,84,29,52,12,53,1,10])
serie1.head(15)
```

Las series por defecto guardan cada valor en un índice numérico que arranca en 0 y va incrementándose. Pero podemos cambiarlo a cualquier tipo de índice el cual no tenga necesariamente un orden (Nota: Los datos están ordenados dentro de la Serie por el orden que están ubicados pero el índice no necesariamente debe tener un orden).

```python
# Indices
serie2 = pd.Series([5.7,8.5,9.1,5.5,8.2,9.0,10,7.0,7.7,9.9],
                   index=["A","B","C","D","E","F","G","H","I","J"])
print(serie2)
```

Las Series se pueden hacer cortes para obtener sub series

```python
# Acceso a los elementos / Filtrado
print(f"El elemento ubicado en la posicion 2 es {serie2[2]}" )
print("La sub serie con los elementos ubicados en la posición 1 y 4 (sin incluir) son:" )
print(serie2[1:4])
```

Además del acceso por ubicación, podemos hacerlo por valor de índice si es que lo tiene

```python
# Acceso a los elementos / Filtrado
print(f"El elemento ubicado en el indice F es {serie2['F']}" )
print("La sub serie con los elementos ubicados en la posición D y G (sin incluir) son:" )
print(serie2["D":"G"])
```

Muy similar a NumPy, las series se pueden manipular con operaciones, siempre y cuando los tipos de datos de las series sean compatibles. A continuación solo hay algunas operaciones de la enorme cantidad que existen. Más operaciones se pueden consultar [aqui](https://pandas.pydata.org/docs/user_guide/dsintro.html#series)

```python
# Operaciones entre series
s1 = pd.Series([1, 3, 5])
s2 = pd.Series([2, 4, 6])
print("Suma:\n", s1 + s2)
print("\nResta:\n", s1 - s2)
print("\nMultiplicación elemento a elemento:\n", s1 * s2)
# Comparaciones
print("\nCada valor es mayor a 3?:\n", s1>3)
# Filtrado
print("\nTodos valores de s1 que son mayores o iguales a 3:\n", s1[s1>=3])

# Calculo estadisticos
print(f"Media de s1: {s1.mean()}")
print(f"STD de s1: {s1.std()}")
```

### Dataframes
Como mencionamos, esta estructura de datos es la más usada de Pandas y nos permite realizar una amplia variedad de operaciones en datos tabulados de forma programática y sencilla.

A continuación vemos como se pueden crear:

```python
# Dataframe vacío
df1 = pd.DataFrame()
df1.head()
```

Los DataFrames tiene columnas con tienen nombres. Además, tienen índices que funcionan exactamente igual que con las Series. Las filas y columnas se pueden acceder por posición y nombre, lo cual veremos más adelante en detalle.

```python
# Data frame desde una lista de listas
l1 = ['Sofía', 33, 'ARG', 1.77]
l2 = ['Peter', 55, 'USA', 1.75]
l3 = ['Juan' , 18, 'CHL', 1.68]
l4 = ['María', 47, 'ECU', 1.50]
l5 = ['Lucas', 80, 'ESP',  1.8]
ListaDeListas = [l1, l2, l3, l4, l5]

df2 = pd.DataFrame(ListaDeListas, columns=['Nombre', 'Edad', 'Origen', 'Altura'])

# El método head nos permite ver las primera 5 filas del DataFrame
df2.head()
```

```python
# Acceso a los elementos / filtrado

print(f"Accediendo por nombre de indice y columna: {df2.loc[0,'Nombre']}") # Nombre de la fila 0
print(f"Accediendo por posición de indice y columna: {df2.iloc[0,0]}") # Nombre de la fila 0

print("\nHaciendo slicing de columnas:")
df2[['Nombre', 'Origen']].head() # Solo me quedo con estas columnas para la visualización
```

Al igual que las series, podemos realizar una infinidad de operaciones con las columnas (cada columna del dataframe es una Serie de Pandas):

```python
# Operaciones entre columnas
df2['2 edades'] = df2['Edad'] * 2 # Agrego columna que es 2 veces la edad
df2.head()
```

```python
df2['NombreYPais'] = df2['Nombre'] + ' de ' + df2['Origen']
df2.head()
```
