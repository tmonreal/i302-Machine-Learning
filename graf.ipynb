```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import triang

# Parámetros normalizados para un gráfico genérico
a_norm = 0
b_norm = 1
c_rel = 0.5  # modo relativo en el centro

# Distribución triangular simétrica
dist = triang(c=c_rel, loc=a_norm, scale=b_norm - a_norm)
x = np.linspace(-0.1, 1.1, 500)
y = dist.pdf(x)

# Crear gráfico
plt.figure(figsize=(8, 4))
plt.plot(x, y, color='black')
plt.fill_between(x, y, alpha=0.2, color='gray')

# Línea vertical en el modo
plt.axvline(x=0.5, color='gray', linestyle='--', label='Modo (c)')

# Eliminar ticks numéricos
plt.xticks([], [])
plt.yticks([], [])

# Posición más baja de etiquetas para que no tapen la curva
label_y = -0.2

# Etiquetas simbólicas
plt.text(0, label_y, r'$a$', ha='center', va='top', fontsize=12)
plt.text(1, label_y, r'$b$', ha='center', va='top', fontsize=12)
plt.text(0.5, label_y, r'$\frac{a + b}{2}$', ha='center', va='top', fontsize=12)

# Ajustes visuales
plt.title('Distribución Triangular Simétrica')
plt.xlabel('')
plt.ylabel('Densidad')
plt.ylim(-0.1, 2.1)
plt.grid(False)
plt.tight_layout()
plt.show()
```
