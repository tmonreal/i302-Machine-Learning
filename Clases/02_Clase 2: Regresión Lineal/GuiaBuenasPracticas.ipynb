### Gr谩ficos simples

Para que un grafico pueda ser interpretado de forma r谩pida y sencilla, el mismo debe cumplir con las siguientes condiciones:

- Todos los <font color='darkmagenta'>ejes</font> deben tener un <font color='darkmagenta'>nombre</font> que defina adecuadamente la magnitud o par谩metro que representa.

- Las <font color='darkmagenta'>unidades</font> de cada magnitud representada debe ser especificada al final del nombre del eje.

- El tama帽o de fuente de los nombres de ejes como de las marcas de graduaci贸n de los mismos deben ser acordes al tama帽o del gr谩fico generado.

- En caso de que el grafico presente m谩s de una serie de datos (por ejemplo, una comparaci贸n de dos curvas de respuesta en frecuencia), se debe realizar una distinci贸n marcada entre ambas curvas, ya sea mediante el uso de diferentes colores o estilos de l铆nea. El c贸digo de representaci贸n utilizado debe ser especificado en un <font color='darkmagenta'>cuadro de leyenda</font>.

- Se recomienda el uso de <font color='darkmagenta'>grilla</font> en los gr谩ficos para poder identificar con mayor facilidad los valores que toma la informaci贸n representada.


> **Ejemplo:** Representaci贸n de se帽ales el茅ctricas senoidales.

````python
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0, 2, 100)
sig1 = np.sin(2*np.pi*t) # Ambas se帽ales tienen una frecuencia de 1Hz.
sig2 = 1+np.sin(2*np.pi*t)

plt.figure(figsize=(7, 5))
plt.plot(t, sig1, color='b', label="Se帽al 1"); # El atributo label define el nombre de la se帽al en la leyenda
plt.plot(t, sig2, color='r',linestyle='--', label="Se帽al 2"); # El atributo linestyle define el trazo de la l铆nea
plt.xlabel("Tiempo [s]", fontsize=14)
plt.ylabel("Voltaje [V]", fontsize=14)
plt.legend(fontsize=12) # Establece el cuadro de leyenda
plt.grid() # Establece la grilla
````
