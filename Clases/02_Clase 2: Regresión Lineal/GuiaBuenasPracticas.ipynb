# Guía de buenas prácticas de programación

- Este documento tiene como finalidad listar una serie de prácticas recomendadas a la hora de escribir código.  
- Estos conceptos son de gran utilidad a la hora de programar de forma clara, prolija y ordenada, lo que permite que otras personas sean capaces de entender de forma rápida y sencilla el funcionamiento del código.
- También, se definen los lineamientos para generar gráficos de forma correcta, con el fin de que estos tengan las cualidades necesarias para que se entiendan facilmente y nos permitan comunicar resultados de manera efectiva.

## 📚 Generalidades

- Las librerías (o mejor dicho bibliotecas) que se implementarán en el código deben ser importadas al inicio del mismo.

- Las funciones empleadas deben ser definidas en su totalidad a continuación de la importación de librerías, antes de empezar a usarlas y de definir variables.

> **Ejemplo:**

````python
# Sección de importación de librerías
import numpy as np

# Sección de definición de funciones:
def funcion_suma(v1, v2):
    
    resultado = a + b
    return resultado

def funcion_resta(v1, v2):
    
    resultado = a - b
    return resultado

# Sección de uso de las funciones creadas:
a = 5
b = -4

resultado_suma = funcion_suma(a, b)
resultado_resta = funcion_resta(a, b)

print("La suma entre {} y {} da como resultado {}".format(a, b, resultado_suma))
print("La resta entre {} y {} da como resultado {}".format(a, b, resultado_resta))
````

## 🔡 Variables

- Las variables son sensibles al uso de mayusculas y minusculas, se recomienda nombrar las variables en su totalidad con <font color='blue'>minusculas</font>, ya que las variables con mayusculas estan reservadas para constantes y clases.

- El <font color='blue'>nombre</font> asignado a una variable debe ser <font color='blue'>representativo del contenido</font> que almacena. No usar variables cuyo nombre carezca de significado descriptivo. Con un código legible y nombres significativos, el código se va-auto documentando.

- Declarar variables en líneas separadas.

- Ser consistente al momento de utilizar un estándar para definir nombres largos. Los formatos utilizados son los siguientes:

    - **camelCase**: Se separan palabras mediante el uso de mayusculas. Ej: totalMediciones
    - **snake_case**: Se separan palabras mediante el uso de guión bajo. Ej: total_Mediciones

> **Ejemplo:**

````python
import numpy as np

presion_medida =  5.43*10**(-3)
presion_ref = 20*10**(-6) # 20μPa es la presión de referencia del sonido en el aire

spl = 20 * np.log10(presion_medida/presion_ref) # Nivel de presión sonora (SPL) medido

print("El nivel de presión sonora medido es de {} dB.".format(spl))
````

## ⚙️ Funciones

- Definir el <font color='green'>nombre</font> de funciones, módulos y clases que mejor <font color='green'>representen las acciones que realizan.</font>

- Al inicio de cada función, <font color='green'>agregar</font> brevemente un <font color='green'>comentario que explique el comportamiento general</font> de la función.

- <font color='green'>Comentar cuando sea justo y necesario,</font> usar los comentarios dentro de las funciones para describir las variables (sólo cuando su utilidad sea potencialmente dudosa) y cuando existan bloques de código difíciles de entender a primera vista; el exceso de comentarios vuelve ilegible el código.

- <font color='green'>Definir variables locales al inicio de la implementación de cada función,</font> como un bloque de código bien separado del bloque que contenga las instrucciones ejecutables, ésta separación puede consistir en una línea en blanco, o bien un comentario que denote la utilidad de cada bloque.

- Evitar la incorporación de más de <font color='green'>una instrucción por línea</font>. Esto reduce notoriamente la legibilidad del código, ya que el programador habitualmente está acostumbrado a leer una instrucción por línea.

- Realizar la <font color='green'>indentación necesaria</font>, si una instrucción abarca más de una línea. Esto permite que la instrucción se entienda con mayor claridad.

> **Ejemplo:**

````python
def ejemplo_suma(a, b):
    """
    Devuelve el resultado de la suma de dos numeros racionales.
    
        Parámetros de entrada:
            a(float): Primer elemento de la suma
            b(float): Segundo elemento de la suma
        
        Parámetros de salida:
            resultado(float): Resultado de la suma de los valores de entrada
    """
    resultado = a + b
    return resultado
````

## 💬 Comentarios y escritura en general
Es una buena idea mantener una coherencia a lo largo de todo el material que desarrollemos. Comúnmente, todo el material se desarrolla en idioma <font color='teal'>inglés</font>. Habran notado que todas las bibliotecas con las que trabajamos utilizan en sus definiciones y documentación el idioma inglés, a pesar de que este no sea el lenguaje nativo de muchas de las personas involucradas en el desarrollo. Es por esto que recomendamos fuertemente adoptar esta metodología a la hora de definir nombres de variables, funciones e incluso también a la hora de escribir documentación. 

Con respecto a los comentarios, vamos a diferenciar **3 tipos de comentarios**. Los primeros dos casos refieren a <font color='teal'>comentarios de párrafo</font>. Comúnmente encontramos estos comentarios en dos secciones. 

1. En primer lugar, al <font color='teal'>inicio de un script que forma parte de un módulo</font> o que requiere alguna información previa a su utilización, ya sea respecto a licencias, a autores, o a particularidades en el uso del módulo. La inclusión de este tipo de comentarios es opcional. Algunos ejemplos:
   - [Ejemplo 1: scipy](https://github.com/scipy/scipy/blob/v1.12.0/scipy/signal/_spectral_py.py#L1047-L1235)
   - [Ejemplo 2: pyfilterbank](https://github.com/SiggiGue/pyfilterbank/blob/master/pyfilterbank/gammatone.py)

2. En segundo lugar, vamos a encontrar comentarios de parrafos como <font color='teal'>documentación de funciones o clases</font>. Estos es, cada vez que implementamos una función o clase, debemos incluir al inicio un comentario de parrafo que contenga:
   - Explicación breve y concisa del propósito de la función/clase.
   - Variables de entrada, especificando tipo y significado de cada una, valores por default, etc.
   - Variables de salida, especificando tipo y significado de cada una. 
    
    Algunos ejemplos:
      - [Ejemplo 1: soundfile read](https://python-soundfile.readthedocs.io/en/0.11.0/_modules/soundfile.html#read) 
      - [Ejemplo 2: librosa load](https://librosa.org/doc/main/_modules/librosa/core/audio.html#load)
        
3. Por último tenemos a los <font color='teal'>comentarios de linea</font>. Estos deben utilizarse para dar alguna aclaración pertinente sobre una secuencia de lineas, sobre alguna variable en particular o para brindar algún tipo de información **en el caso en que no alcance con leer el código para comprender lo necesario**. Idealmente, un código bien escrito y estructurado no deberia necesitar de este tipo de comentarios, dejandose mas para un uso en etapa de desarrollo.

Por último, pero no menos importante, se debe tener precaución con los **saltos de linea** dentro de un mismo script. Al principio, es común que intentemos utilizar el recurso de salto de linea para sumar claridad al código. Lo cierto es que un código bien escrito y estructurado no se beneficia de excesivos saltos de linea. Normalmente, dejamos los saltos de lineas para <font color='teal'>separar los imports del resto del script, la definición de una función, o la definición de una clase.</font> Por ejemplo, podemos usar un salto de linea para separar los imports del resto del script, un salto de linea como separación entre funciones, y un doble salto de linea para separación de clases. Idealmente, no debería ser necesario utilizar un salto de linea en ninguna otra ocasión.

## 💾 Versiones finales
Por versión final entendemos algún punto (inicial, intermedio o final) de la etapa de desarrollo de código en la cual <font color='olive'>guardamos una versión de nuestro trabajo, la publicamos o simplemente la compartimos</font> con alguien más, ya sea para colaborar o para pedir una devolución critica. Antes de finalizar estas versiones finales, debemos asegurarnos de que nuestro codigo cumpla con algunos lineamientos:
- <font color='crimson'>No debe contener comentarios de desarrollo.</font> Por ejemplo:
 
    ```python
    #TODO: apply samplerate normalization
    ```
- <font color='crimson'>No debe contener lineas de código comentadas.</font> Si esta comentado, son lineas que no se van a ejecutar, entonces se deberian poder eliminar. Por ejemplo:

    ```python
    #import librosa
    import soundfile as sf
    ```
    ```python
    def load_audio(path):
        #signal, sr = librosa.load(path, sr=None)
        signal, sr = sf.read(path)
        return signal, sr
    ```
- <font color='crimson'>No debe generar ningún tipo de warnings</font>, por mas que estos no impidan la ejecución del script. 

## 📈 Gráficos

### Gráficos simples

Para que un grafico pueda ser interpretado de forma rápida y sencilla, el mismo debe cumplir con las siguientes condiciones:

- Todos los <font color='darkmagenta'>ejes</font> deben tener un <font color='darkmagenta'>nombre</font> que defina adecuadamente la magnitud o parámetro que representa.

- Las <font color='darkmagenta'>unidades</font> de cada magnitud representada debe ser especificada al final del nombre del eje.

- El tamaño de fuente de los nombres de ejes como de las marcas de graduación de los mismos deben ser acordes al tamaño del gráfico generado.

- En caso de que el grafico presente más de una serie de datos (por ejemplo, una comparación de dos curvas de respuesta en frecuencia), se debe realizar una distinción marcada entre ambas curvas, ya sea mediante el uso de diferentes colores o estilos de línea. El código de representación utilizado debe ser especificado en un <font color='darkmagenta'>cuadro de leyenda</font>.

- Se recomienda el uso de <font color='darkmagenta'>grilla</font> en los gráficos para poder identificar con mayor facilidad los valores que toma la información representada.


> **Ejemplo:** Representación de señales eléctricas senoidales.

````python
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0, 2, 100)
sig1 = np.sin(2*np.pi*t) # Ambas señales tienen una frecuencia de 1Hz.
sig2 = 1+np.sin(2*np.pi*t)

plt.figure(figsize=(7, 5))
plt.plot(t, sig1, color='b', label="Señal 1"); # El atributo label define el nombre de la señal en la leyenda
plt.plot(t, sig2, color='r',linestyle='--', label="Señal 2"); # El atributo linestyle define el trazo de la línea
plt.xlabel("Tiempo [s]", fontsize=14)
plt.ylabel("Voltaje [V]", fontsize=14)
plt.legend(fontsize=12) # Establece el cuadro de leyenda
plt.grid() # Establece la grilla
````

### Gráficos compuestos

En el caso de presentar un conjunto de dos o más gráficos en simultaneo que mantengan relación entre si, los siguientes lineamientos son de gran ayuda para que resulte más sencillo analizarlos de forma correcta.

- Se debe tratar de <font color='darkmagenta'>mantener la escala en todos los gráficos</font> (tanto en el eje X como en Y) para facilitar la comparación entre gráficos.

- Los títulos de los sub-gráficos deben especificar que se representa en cada uno de ellos.

- Si se representa más de una señal en los diferentes gráficos, <font color='darkmagenta'>utilizar el mismo código de representación</font> (color y trazo de línea) en cada uno de ellos.

- En el caso de que los gráficos representen una misma magnitud en un determinado rango, se puede compartir el nombre de eje (ver eje X en el ejemplo siguiente).

> **Ejemplo:** Representación de señales eléctricas de entrada y salida de un sistema.

````python
x = np.linspace(0, 2, 500)
y = 2* np.sin(x **5)

fig, (ax1, ax2) = plt.subplots(nrows=2, figsize=(8,6), sharey=True)

ax1.plot(x, y, label="Caso 1")
ax1.set_title("Señal de entrada", fontsize=15)
ax1.set_xlim(0,2)
ax1.set_ylabel("Voltaje [V]", fontsize=14)
ax1.grid()

ax2.plot(x, -y, color="r")
ax2.set_xlabel("Tiempo [s]", fontsize=14)
ax2.set_title("Señal de salida", fontsize=15)
ax2.set_xlim(0,2)
ax2.set_ylabel("Voltaje [V]", fontsize=14)
ax2.grid()

fig.tight_layout()
````
