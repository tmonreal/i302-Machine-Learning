```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función y su transformación
x = np.linspace(0, 4, 400)
f = - (x - 2)**2 + 4        # f(x) tiene máximo en x=2, f(2)=4
f_neg = -f                  # -f(x) tiene mínimo en x=2, -f(2)=-4

fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# Gráfico izquierdo: f(x)
axs[0].plot(x, f, label=r'$f(x) = - (x-2)^2 + 4$')
axs[0].plot(2, 4, 'ro', label='Máximo')
axs[0].set_title('Maximización de f(x)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('f(x)')
axs[0].legend()
axs[0].grid(True)

# Gráfico derecho: -f(x)
axs[1].plot(x, f_neg, label=r'$-f(x)$')
axs[1].plot(2, -4, 'go', label='Mínimo')
axs[1].set_title('Minimización de -f(x)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('-f(x)')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir f(x) y su transformación g(f(x)) = sqrt(f(x))
x = np.linspace(0, 6, 400)
f = (x - 3)**2          # f(x) tiene mínimo en x = 3
g_f = np.sqrt(f)        # g es la función raíz cuadrada

# Para demostrar que sqrt es monótona creciente, graficamos sqrt(z) para z en [0, max(f)]
z = np.linspace(0, np.max(f), 400)
sqrt_z = np.sqrt(z)

fig, axs = plt.subplots(1, 3, figsize=(18, 4))

# Subplot 1: Función f(x)
axs[0].plot(x, f, label=r'$f(x)=(x-3)^2$')
axs[0].plot(3, 0, 'ro', label='Mínimo')
axs[0].set_title('Función f(x)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('f(x)')
axs[0].legend()
axs[0].grid(True)

# Subplot 2: Transformación g(f(x)) = sqrt(f(x))
axs[1].plot(x, g_f, label=r'$g(f(x))=\sqrt{f(x)}$')
axs[1].plot(3, 0, 'ro', label='Mínimo')
axs[1].set_title('Transformación: g(f(x))')
axs[1].set_xlabel('x')
axs[1].set_ylabel(r'$g(f(x))$')
axs[1].legend()
axs[1].grid(True)

# Subplot 3: Gráfico de la función raíz cuadrada, demostrando que es monótonamente creciente
axs[2].plot(z, sqrt_z, label=r'$\sqrt{z}$')
axs[2].set_title('Función $\sqrt{z}$: Monótonamente Creciente')
axs[2].set_xlabel('z')
axs[2].set_ylabel(r'$\sqrt{z}$')
axs[2].legend()
axs[2].grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función f(x) y su logaritmo
x = np.linspace(0, 4, 400)
f = (x - 2)**2 + 1        # f(x) tiene mínimo en x=2, f(2)=1 (siempre positivo)
log_f = np.log(f)

fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# Gráfico izquierdo: f(x)
axs[0].plot(x, f, label=r'$f(x) = (x-2)^2 + 1$')
axs[0].plot(2, 1, 'ro', label='Mínimo')
axs[0].set_title('Función f(x)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('f(x)')
axs[0].legend()
axs[0].grid(True)

# Gráfico derecho: log(f(x))
axs[1].plot(x, log_f, label=r'$\log(f(x))$')
axs[1].plot(2, np.log(1), 'ro', label='Mínimo')
axs[1].set_title('Función log(f(x))')
axs[1].set_xlabel('x')
axs[1].set_ylabel('log(f(x))')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la función: f(x, y) = x² + y²
def f(x, y):
    return x**2 + y**2

# Creamos una malla de puntos para graficar las líneas de contorno
x = np.linspace(-3, 3, 400)
y = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# La derivada de f es: grad_f = [2x, 2y]
def grad_f(x, y):
    return 2*x, 2*y

# Elegimos un punto de interés, por ejemplo (1,2)
x0, y0 = 1, 2
grad_x, grad_y = grad_f(x0, y0)

# Calculamos un vector tangente a la línea de contorno en (1,2)
# Se obtiene rotando el gradiente 90°: (–grad_y, grad_x)
tan_x, tan_y = -grad_y, grad_x

# Creamos la figura y graficamos las líneas de contorno de f(x, y)
plt.figure(figsize=(6,6))
contours = plt.contour(X, Y, Z, levels=10, cmap='viridis')
plt.clabel(contours, inline=True, fontsize=8)

# Marcamos el punto de interés
plt.plot(x0, y0, 'ro', label='Punto (1,2)')

# Dibujamos el vector gradiente en el punto (1,2)
plt.quiver(x0, y0, grad_x, grad_y, angles='xy', scale_units='xy', scale=1, color='red', label='Gradiente f')

# Dibujamos el vector tangente a la línea de contorno en el mismo punto
plt.quiver(x0, y0, tan_x, tan_y, angles='xy', scale_units='xy', scale=1, color='blue', label='Dirección tangente')

plt.title('El gradiente es perpendicular a la línea de contorno')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Definir el rango de x y las funciones f(x) y g(x)
x = np.linspace(-1, 3, 200)
f = x**2
g = (x - 2)**2 + 1

# 2. Hallar los mínimos individuales de cada función (solo para marcarlos en la gráfica)
x_min_f = 0   # mínimo de f(x) = x^2
x_min_g = 2   # mínimo de g(x) = (x-2)^2 + 1

f_min = x_min_f**2
g_min = (x_min_g - 2)**2 + 1

# 3. Crear figura con 2 subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# --- Subplot 1: Gráfica de f(x) y g(x) ---
axs[0].plot(x, f, label='f(x) = x^2', color='red')
axs[0].plot(x, g, label='g(x) = (x - 2)^2 + 1', color='blue')

# Marcar los mínimos
axs[0].plot(x_min_f, f_min, 'ro', label='Mínimo f')
axs[0].plot(x_min_g, g_min, 'bo', label='Mínimo g')

axs[0].set_title('Funciones objetivo')
axs[0].set_xlabel('x')
axs[0].set_ylabel('Valor de la función')
axs[0].legend()
axs[0].grid(True)

# --- Subplot 2: Curva (f(x), g(x)) en el espacio de objetivos ---
axs[1].plot(f, g, color='purple', label='(f(x), g(x))')

# Marcar los puntos correspondientes a x=0 y x=2
axs[1].plot(f_min, g[x == x_min_f], 'ro', label='Punto min f')
axs[1].plot(f[x == x_min_g], g_min, 'bo', label='Punto min g')

axs[1].set_title('Espacio de objetivos (f, g)')
axs[1].set_xlabel('f(x)')
axs[1].set_ylabel('g(x)')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos un parámetro t entre 0 y 1
t = np.linspace(0, 1, 100)

# Funciones objetivo (parametrizadas)
f = t**2
g = (1 - t)**2

# (Opcional) Generamos algunos puntos "aleatorios" para ilustrar que
# no todos están en la frontera de Pareto
n_puntos = 20
f_rand = np.random.rand(n_puntos)**2
g_rand = np.random.rand(n_puntos)**2

# Creamos la figura con 2 subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Subplot 1: f(t) y g(t) en función de t
axs[0].plot(t, f, label='f(t) = t²', color='red')
axs[0].plot(t, g, label='g(t) = (1 - t)²', color='blue')
axs[0].set_title('Funciones objetivo')
axs[0].set_xlabel('t')
axs[0].set_ylabel('Valor de la función')
axs[0].grid(True)
axs[0].legend()

# Subplot 2: Espacio de objetivos (f, g)
axs[1].plot(f, g, 'r-', label='Frente de Pareto')
axs[1].scatter(f_rand, g_rand, color='blue', alpha=0.6, label='Otras soluciones')
axs[1].set_title('Espacio de objetivos (f, g)')
axs[1].set_xlabel('f(t)')
axs[1].set_ylabel('g(t)')
axs[1].grid(True)
axs[1].legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Punto óptimo y gradientes calculados
x0, y0 = 0.8, 0.2
grad_f = np.array([2*x0, 8*y0])   # [1.6, 1.6]
grad_h = np.array([1, 1])          # [1, 1]
lambda_val = 1.6  # Escalar que relaciona ambos gradientes

plt.figure(figsize=(6,6))

# Graficamos la línea de restricción h(x,y)= x+y-1=0
x_vals = np.linspace(0, 1.5, 100)
y_vals = 1 - x_vals
plt.plot(x_vals, y_vals, label='Restricción: x+y=1', color='blue')

# Graficamos una línea de nivel de f que pase por el óptimo (f=0.8^2+4*(0.2)^2=0.8)
X, Y = np.meshgrid(np.linspace(0, 1.5, 100), np.linspace(0, 1.5, 100))
F = X**2 + 4*Y**2
contour = plt.contour(X, Y, F, levels=[0.8], colors='green')
plt.clabel(contour, inline=True, fontsize=10, fmt='f=0.8')

# Marcamos el punto óptimo
plt.plot(x0, y0, 'ro', label='Óptimo (0.8, 0.2)')

# Dibujamos el vector gradiente de f (en rojo) y de h (en púrpura) en el óptimo
plt.quiver(x0, y0, grad_f[0], grad_f[1], angles='xy', scale_units='xy', scale=1, color='red', label=r'$\nabla f$')
plt.quiver(x0, y0, grad_h[0], grad_h[1], angles='xy', scale_units='xy', scale=1, color='purple', label=r'$\nabla h$')

# Configuramos el gráfico
plt.xlim(0, 2.5)
plt.ylim(0, 2.5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Multiplicadores de Lagrange: Gradientes en el óptimo')
plt.legend()
plt.grid(True)

# Anotamos la relación ∇f = λ ∇h
plt.annotate(r'$\nabla f = \lambda\, \nabla h$', xy=(x0 + grad_h[0]/2, y0 + grad_h[1]/2), 
             xytext=(x0+0.5, y0+1.0),
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3"))

plt.show()
```
